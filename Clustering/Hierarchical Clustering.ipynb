## HIERARCHICAL CLUSTERING

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Mall_Customers.csv",index_col=0)
```

```python
df.head()
```

```python
datasubset=df.loc[:,["Age","Annual Income (k$)"]]
#if i want to use iloc i had have to replace the name of column to there index no. 
```

```python
plt.figure(figsize=(10,7))
plt.scatter(datasubset[["Age"]],datasubset[["Annual Income (k$)"]],s=50,c="purple")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
#creating an dendrogram to see the branches and to visualise the connnection 
dendrogram=sch.dendrogram(sch.linkage(datasubset.head(50),method="ward"))
plt.title("DENDOGRAM")
plt.xlabel("Age of customers")
plt.ylabel("Annual Income")
```

-  Here we have to check for the longest line and the no. og  line comes under that longest line is the optimla no. of cluster 
-  as for this case only <b> two cluster could be made</b>

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage="ward")
cluster.fit_predict(datasubset)
```

```python
cl=cluster.fit_predict(datasubset)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(datasubset,cl)
```
